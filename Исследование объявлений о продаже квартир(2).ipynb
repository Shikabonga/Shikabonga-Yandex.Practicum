## Предобработка данных

### Названия столбцов

```python
#Начнем с названия столбцов
columns_name_new = [
    'images_total', #кол-во фотографий в объявлении
    'last_price', #цена на момент снятия с публикации
    'area_total_m', #площадь квартиры в квадратных метрах (м²)
    'date_of_publication', #дата публикации
    'rooms_count', #кол-во комнат
    'ceiling_height_m', #высота потолков (м)
    'floors_total', #всего этажей в доме
    'living_area_m', #жилая площадь в квадратных метрах(м²)
    'floor_number', #этаж
    'is_apartment', #апартаменты
    'is_studio', #квартира-студия
    'is_open_plan', #свободная планировка
    'kitchen_area_m', #площадь кухни в квадратных метрах (м²)
    'balconies_count', #число балконов
    'city_name', #название населённого пункта
    'airports_nearest_m', #расстояние до ближайшего аэропорта в метрах (м)
    'centers_nearest_m', #расстояние до центра города (м)
    'parks_around_3km', #число парков в радиусе 3 км
    'parks_nearest_m', #расстояние до ближайшего парка (м)
    'ponds_around_3km', #число водоёмов в радиусе 3 км
    'ponds_nearest_m', #расстояние до ближайшего водоёма (м)
    'days_exposition' #сколько дней было размещено объявление (от публикации до снятия)
]
data.set_axis(columns_name_new, axis = 'columns', inplace = True) #заменяем
data.columns #проверка
```

### Меняем порядок

```python
data = data[[
    'city_name',
    'is_studio',
    'is_apartment',
    'rooms_count',
    'last_price',
    'area_total_m',
    'date_of_publication',
    'living_area_m',
    'kitchen_area_m',
    'floor_number',
    'floors_total',
    'balconies_count',
    'is_open_plan',
    'ceiling_height_m',
    'images_total',
    'airports_nearest_m',
    'centers_nearest_m',
    'parks_around_3km',
    'parks_nearest_m',
    'ponds_around_3km',
    'ponds_nearest_m',
    'days_exposition'
]]
data.columns #проверка
```

### Работа с пропусками

```python
print('Кол-во nan значений в названиях НП:',data['city_name'].isna().sum()) #кол-во пропусков в названиях НП
data = data.dropna(subset=['city_name']) #удаляем строки где нет названия НП, эти данные бесполезны для анализа
print('Кол-во nan значений в названиях НП после удлаения:',data['city_name'].isna().sum()) #проверяем
print('Кол-во уникальных названий НП до замены:',len(data['city_name'].unique())) #кол-во уникальных названий НП
data['city_name'] = data['city_name'].str.replace('ё','е') #замена в названии НП "ё" на "е"
print('Кол-во уникальных названий НП после замены:',len(data['city_name'].unique())) #проверяем
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False) #уберем пропуски из 'is_apartament', скорее всего отсутствует признак апартаментов
data['is_apartment'].unique() #проверка
```

### Перевод значений в целочисленый тип 

```python
#Перет сменой типа, проверим что там лежит
def columns_is_nan(data,columns):
    for col in columns:
        print('Nan знфчений в {}:{}'.format(col,data[col].isna().sum()))

def float_to_int(data,columns): #функция для перевода в тип int списка столбцов из DataFrame. Принимает значения: data - DataFrame, columns - список столбцов
    for col in columns:
        data[col] = data[col].astype('Int64',errors='ignore')

#data.info()
columns = [ #формируем список
    'last_price', #в цене редко есть копейки
    'floors_total', #не может быть 2.5 этажа в доме, в пригодном для жизни по крайней мере
    'balconies_count', #тоже самое и с балконами
    'days_exposition' #дни для кучи тоже переведем, для красоты
]
columns_is_nan(data,columns) #проверим список на nan
```

```python
data['balconies_count'] = data['balconies_count'].fillna(0)#nan значения заменим на 0, скорее всего,если балконов нет, то и не заполняли
data['floors_total'] = data['floors_total'].fillna(0)#nan значений не много, на результат не повлияет, заменим на 0

float_to_int(data,columns) #переводим, то что переведется
data.info()#проверяем
```

Оставшиеся пропуски, либо малая часть, либо остутствие в черте НП(парки, водоемы, аэропорты), либо данные которые нам не понадобятся для исследования.

## Расчёты и добавление результатов в таблицу

```python
data['price_per_metr'] = data['last_price']/data['area_total_m']#расчет цены за м²
data['price_per_metr'] = data['price_per_metr'].round(decimals=2)#округление до 2х знаков
data['date_of_publication'] = pd.to_datetime(data['date_of_publication'], format = '%Y-%m-%d')
data['day_of_week_of_publication'] = data['date_of_publication'].dt.day_name() #столбец с днем недели
data['month_of_publication'] = data['date_of_publication'].dt.month_name() #столбец с месяцем
data['year_of_publication'] = data['date_of_publication'].dt.year #столбец с годом
data['centers_nearest_km'] = data['centers_nearest_m']/1000 #добавим столбец расстояние до центра в км
data['last_price_million'] = data['last_price']/1000000 #добавим столбец цены в миллионах, для удобного просмотра

def floor_status(row): #функция для назначения категории этажу, на входе строка
    if row['floor_number'] == row['floors_total']:
        return 'последний'
    elif row['floor_number'] == 1:
        return 'первый'
    else:
        return 'другой'
        
data['floor_status'] = data.apply(floor_status, axis = 1)#разобъем этажи на категории
data['ratio_living_total_area'] = data['living_area_m']/data['area_total_m'] #соотношение жилой и общей площади
data['ratio_living_total_area'] = data['ratio_living_total_area'].round(decimals = 2)
data['ratio_kitchen_total_area'] = data['kitchen_area_m']/data['area_total_m'] #отношение площади кухни к общей
data['ratio_kitchen_total_area'] = data['ratio_kitchen_total_area'].round(decimals = 2)
data.head(10)
```

```python
def clear_data(df,col): #функция для удаления вбросов
    q1 = df[col].quantile(0.25) #1 квартиль
    q3 = df[col].quantile(0.75) #3 квартиль
    iqr = q3-q1 #межквартильный размах
    return df.query(col+' > @q1-1.5*@iqr & '+col+' < @q3+1.5*@iqr')

stat_data = clear_data(data,'area_total_m') #чистим от вбросов по столбцу площади
stat_data = clear_data(stat_data,'rooms_count') #чистим от вбросов по кол-ву комнат
stat_data = clear_data(stat_data,'ceiling_height_m') #чистим от вбросов по высоте потолков

stat_data['area_total_m'].hist(bins = 100, range=(12,500),figsize=(7,7))
plt.xlabel("Площадь, м²")
plt.ylabel("Кол-во объявлений")
plt.title("Гистограмма по общей площади");
plt.show()
stat_data['last_price_million'].hist(bins = 100, range=(0,80),figsize=(7,7))
plt.xlabel("Цена, млн.")
plt.ylabel("Кол-во объявлений")
plt.title("Гистограмма по цене");
plt.show()
stat_data['rooms_count'].value_counts().plot(kind='bar')
plt.xlabel("Кол-во комнат")
plt.ylabel("Кол-во объявлений")
plt.title("Гистограмма по кол-ву комнат");
plt.show()
stat_data['ceiling_height_m'].hist(bins = 50, range=(2,5),figsize=(7,7))
plt.xlabel("Высота потолков, м.")
plt.ylabel("Кол-во объявлений")
plt.title("Гистограмма по высоте потолков");
plt.show()
```
